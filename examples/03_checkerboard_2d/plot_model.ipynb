```python
import pygmt
import h5py
import numpy as np
```

```python
with h5py.File('OUTPUT_FILES/final_model.h5', 'r') as f:
    svel = f['vel_ph'][:]
    x = f['x'][:]
    y = f['y'][:]
    periods = f['periods_ph'][:]
xx, yy = np.meshgrid(x, y)
```

```python
num_period = 5
region=[x.min(), x.max(), y.min(), y.max()]
fig = pygmt.Figure()
fig.basemap(region=region, projection='M4i', frame=True)
grid = pygmt.surface(z=svel[:,:,num_period].ravel(), x=xx.ravel(),y=yy.ravel(), region=region, spacing='0.01/0.01')
fig.grdimage(grid=grid, cmap='seis')
fig.show()
```
